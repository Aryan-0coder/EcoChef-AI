```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Load model
model = SentenceTransformer(
    "sentence-transformers/all-MiniLM-L6-v2"
)

# Your chunks
chunks = [
    "Tomatoes should be stored at room temperature and consumed within 5 days.",
    "Potatoes should be stored in a cool, dry, dark place.",
    "Spinach should be refrigerated and consumed within 3 days.",
    "Aloo Curry: Ingredients: Potato, Tomato, Onion",
    "Vegetable Soup: Ingredients: Carrot, Tomato, Cabbage",
    "Food waste contributes to methane emissions.",
    "Reducing food waste helps lower carbon emissions and saves resources."
]

import chromadb

client = chromadb.PersistentClient(path="./chroma_db")

collection = client.get_collection(
    name="food_knowledge"
)

query = input("Ask a question: ")

results = collection.query(
    query_texts=[query],
    n_results=3
)

retrieved_chunks = results["documents"][0]

retrieved_chunk = "\n\n".join(retrieved_chunks)

print("\nQuestion:")
print(query)

print("\nRetrieved Context:")
print(retrieved_chunk)

prompt = f"""
Use the context below to answer the user's question completely.

Context:
{retrieved_chunk}

Question:
{query}

Provide a detailed answer with numbered points.
"""
from ibm_watsonx_ai import Credentials
from ibm_watsonx_ai.foundation_models import ModelInference
from dotenv import load_dotenv
import os

load_dotenv()

credentials = Credentials(
    url=os.getenv("URL"),
    api_key=os.getenv("API_KEY")
)

model = ModelInference(
    model_id="ibm/granite-4-h-small",
    credentials=credentials,
    project_id=os.getenv("PROJECT_ID"),
    params={
        "max_new_tokens": 250,
        "temperature": 0.3
    }
)

response = model.generate_text(
    prompt=prompt
)

print("\nGranite Answer:\n")
print(response)
```
